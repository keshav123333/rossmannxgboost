```python
ross_df=pd.read_csv("/content/rossmann-store-sales/train.csv")
store_df=pd.read_csv("/content/rossmann-store-sales/store.csv");
test_df=pd.read_csv("/content/rossmann-store-sales/test.csv");
submision_df=pd.read_csv("/content/rossmann-store-sales/sample_submission.csv");
```

```python
merged_df=ross_df.merge(store_df, on="Store",how="left")
merged_test_df=test_df.merge(store_df, on="Store",how="left")
merged_df.info()
```

```python
def spilt_data(df):
  df["date"]=pd.to_datetime(df["Date"])
  df["year"]=df["date"].dt.year
  df["month"]=df["date"].dt.month
  df["day"]=df["date"].dt.day
  df["weekofyear"]=df.date.dt.isocalendar().week;
```

```python
spilt_data(merged_df)
spilt_data(merged_test_df)
```

```python
merged_df=merged_df[merged_df.Open==1].copy()
```

```python
def comp_months(df):
    df["CompetitionOpen"] = (
        12 * (df["year"] - df["CompetitionOpenSinceYear"]) +
        (df["month"] - df["CompetitionOpenSinceMonth"])
    )
    df["CompetitionOpen"] = df["CompetitionOpen"].map(lambda x: 0 if x < 0 else x).fillna(0)
    return df

merged_df = comp_months(merged_df)
merged_test_df = comp_months(merged_test_df)
```

```python
import numpy as np
def check_promo_month(row):
    try:
        month2str = {
            1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
            7: 'Jul', 8: 'Aug', 9: 'Sept', 10: 'Oct', 11: 'Nov', 12: 'Dec'
        }

        months = (row.get('PromoInterval') or '').split(',')

        if row.get('Promo2Open') and month2str.get(row.get('Month')) in months:
            return 1
        else:
            return 0
    except Exception:
        return 0

def promo_cols(df):
    # Calculate months since Promo2 was open
    df['Promo2Open'] =12* (( df['year']) -df['Promo2SinceYear'] ) + (df['Promo2SinceWeek'] - df['weekofyear'])/4


    # Ensure Promo2Open is non-negative and handle NaN values
    df['Promo2Open'] = df['Promo2Open'].apply(lambda x: 0 if x<0 else x).fillna(0)* df['Promo2']

    # Check if the current month is a promotional month and multiply by 'Promo2' column
    df['IsPromo2Month'] = df.apply(check_promo_month, axis=1) * df['Promo2']

    return df

```

```python
promo_cols(merged_df)
promo_cols(merged_test_df)
```

```python
input_cols = [
    'Store', 'DayOfWeek', 'Promo', 'StateHoliday', 'SchoolHoliday',
    'StoreType', 'Assortment', 'CompetitionDistance', 'CompetitionOpen',
    'day', 'month', 'year', 'weekofyear', 'Promo2', 'Promo2Open', 'IsPromo2Month'
]

target_col = 'Sales'
```

```python
inputs=merged_df[input_cols].copy()
targets=merged_df[target_col].copy()
```

```python
numeric_cols = [
    'Store', 'Promo', 'SchoolHoliday', 'CompetitionDistance', 'CompetitionOpen',
    'Promo2', 'Promo2Open', 'IsPromo2Month', 'day', 'month', 'year', 'weekofyear'
]

categorical_cols = ['DayOfWeek', 'StateHoliday', 'StoreType', 'Assortment']
```

```python
inputs[numeric_cols].isna().sum()
test_inputs=merged_test_df[input_cols].copy()
```

```python
max_dist=inputs["CompetitionDistance"].max()
inputs["CompetitionDistance"]=inputs["CompetitionDistance"].fillna(2*max_dist,inplace=True)
test_inputs["CompetitionDistance"]=test_inputs["CompetitionDistance"].fillna(2*max_dist,inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(inputs[numeric_cols])
inputs[numeric_cols]=scaler.transform(inputs[numeric_cols])
test_inputs[numeric_cols]=scaler.transform(test_inputs[numeric_cols])


inputs[categorical_cols]
```

```python
from sklearn.preprocessing import OneHotEncoder

# Ensure all categorical values are strings
inputs[categorical_cols] = inputs[categorical_cols].astype(str)
test_inputs[categorical_cols] = test_inputs[categorical_cols].astype(str)

# Initialize the encoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit and transform the categorical columns
encoder.fit(inputs[categorical_cols])
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))

# Apply the transformation
inputs[encoded_cols] = encoder.transform(inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
X=inputs[numeric_cols+encoded_cols]
X_test=test_inputs[numeric_cols+encoded_cols]
```

```python

from xgboost import XGBRegressor
model=XGBRegressor(n_estimators=20,learning_rate=0.05,n_jobs=-1,random_state=42)
model.fit(X,targets)
pred=model.predict(X)

```

```python
from sklearn.metrics import mean_squared_error

def rmse(pred,targets):
 mse = mean_squared_error(targets, pred )


 return mse**0.5

```

```python
model=XGBRegressor(n_estimators=1000,learning_rate=0.2123,n_jobs=-1,random_state=42,max_depth=10,subsample=0.9,colsample_bytree=0.9)
model.fit(X,targets)
pred=model.predict(X)
rmse(pred,targets)
```
